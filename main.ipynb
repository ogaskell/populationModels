```python
%matplotlib widget 

import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
```

```python
BETA = .00001
GAMMA = 0.05
DELTA = .00001
```

```python
def newPop(currentPop, crabPop=0, maxPop=100000, k=1):
    changeInPop = k*currentPop*(1-(currentPop/maxPop))
    Ax = currentPop + changeInPop
    Bxy = currentPop * crabPop * BETA
    return Ax - Bxy

def newCrabPop(currentPop, musselPop):
    change = (DELTA * currentPop * musselPop) - (GAMMA * currentPop)
    return currentPop + change
```

```python
def get99k(y):
    for n in y:
        if n > 99000:
            return y.index(n)
```

```python
x = list(range(1000))
K = 0.3
musselPops = [200]
crabPops = [0]
for n in range(len(x)-1):
    musselPops.append(newPop(musselPops[-1],crabPop=crabPops[-1],k=K))
    if n < 100:
        crabPops.append(0)
    elif n == 100:
        crabPops.append(100)
    else:
        crabPops.append(newCrabPop(crabPops[-1],musselPops[-2]))
```
